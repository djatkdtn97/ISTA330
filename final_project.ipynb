```python
def get_datasets():
    # Location where Data is saved on the net
    data = "https://www.kaggle.com/datasets/rupakroy/waiter-tips-dataset-for-prediction"
    # Downlaod the data
    od.download(data)
    # The data will be saved in the waiter-tips-dataset-for-prediction folder
    data_dir = './waiter-tips-dataset-for-prediction'
    # Accessing the two csv files
    tips_df = pd.read_csv('waiter-tips-dataset-for-prediction/tips.csv')
    return tips_df
tips_df = get_datasets()
```

```python
sns.relplot(x="total_bill", y="tip", hue="smoker", style="time", data=tips_df);
plt.title("Tip amount of Smoker/Non Smoker depends on time", size=15)
plt.xlabel('Tip Amount', fontsize=15, fontweight='bold' )
plt.ylabel('Total Bill Amount', fontsize=15, fontweight='bold')
plt.show()
```

```python
a = sns.relplot(x="total_bill", y="tip", hue="sex", kind = 'line', col="time", data=tips_df)
a.fig.suptitle('Tip depends on Gender')
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips_df);
plt.title("Tip amount of Smoker/Non Smoker", size=20)
plt.xlabel('Tip Amount', fontsize=15, fontweight='bold' )
plt.ylabel('Total Bill Amount', fontsize=15, fontweight='bold')
plt.show()
```
